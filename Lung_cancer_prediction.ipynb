```python
import numpy as np
import pandas as pd

# import the data required
data = pd.read_csv('/content/survey lung cancer.csv')
print(data.head())
```

```python
print(data['GENDER'].unique())
print(data['LUNG_CANCER'].unique())
```

```python

# Import label encoder
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
data['GENDER']= label_encoder.fit_transform(data['GENDER'])

data['GENDER'].unique()

data['LUNG_CANCER']= label_encoder.fit_transform(data['LUNG_CANCER'])

data['LUNG_CANCER'].unique()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pgmpy.models import BayesianModel
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
data = data
heart_disease = pd.DataFrame(data)
print(heart_disease)

model = BayesianModel([
    ('ALCOHOL CONSUMING', 'LUNG_CANCER'),
    ('GENDER', 'LUNG_CANCER'),
    ('SMOKING', 'LUNG_CANCER'),
    ('LUNG_CANCER', 'YELLOW_FINGERS'),
    ('LUNG_CANCER', 'CHEST PAIN')

])

model.fit(heart_disease, estimator=MaximumLikelihoodEstimator)
HeartDisease_infer = VariableElimination(model)
print('For  enter ALCOHOL CONSUMING 2,1 ')
print('For Gender enter Male:1, Female:0')
print('For CHEST PAIN Yes:1, No:0')
print('For SMOKING 1,2')
print('for yellow fingers 1,2')


q = HeartDisease_infer.query(variables=['LUNG_CANCER'], evidence={
    'ALCOHOL CONSUMING': int(input('Enter ALCOHOL CONSUMING: ')),
    'GENDER': int(input('Enter Gender: ')),
    'CHEST PAIN': int(input('Enter CHEST PAIN: ')),
    'SMOKING': int(input('Enter SMOKING: ')),
    'YELLOW_FINGERS': int(input('Enter YELLOW_FINGERS: ')),

})

print(q)
```
